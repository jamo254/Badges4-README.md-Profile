## Function to count number of parameters in each model

```python
def model_size(model):
  return sum(t.numel() for t in model.parameters())

print(f"GPT  size: {model_size(generation_gpt.model)/1000**2:.1f}M parameters")

print(f"GPT2 size: {model_size(generation_gpt2.model)/1000**2:.1f}M parameters")

```

### Generation output from user prompt, produce completions

```python
def enum_pipeline_ouputs(pipe, prompt, num_return_sequences):
    out = pipe(prompt, num_return_sequences=num_return_sequences,
               clean_up_tokenization_spaces=True)
    return "\n".join(f"{i+1}." + s["generated_text"] for i, s in enumerate(out))

prompt = "\nTell me you"
print("GPT completions:\n" + enum_pipeline_ouputs(generation_gpt, prompt, 3))
print("")
print("GPT-2 completions:\n" + enum_pipeline_ouputs(generation_gpt2, prompt, 3))

```

```python
from tqdm.auto import tqdm
from datasets import load_dataset

# Size of document
length = 200000
batch_size = 32  # Define the batch size here
dataset_name = 'transformersbook/codeparrot'
dataset = load_dataset(dataset_name, split="train", streaming=True)
iter_dataset = iter(dataset)

def batch_iterator():
    for _ in tqdm(range(0, length, batch_size)):
        yield [next(iter_dataset)['content'] for _ in range(batch_size)]

new_tokenizer = tokenizer.train_new_from_iterator(batch_iterator(), vocab_size=32768, initial_alphabet=base_vocab)
```

```python
# Sort tokens by their BPE ID in descending order.
tokens = sorted(new_tokenizer.vocab.items(), key=lambda x: x[1], reverse=True)

# Extract only the token strings, not their IDs.
token_strings = [token for token, _ in tokens[257:280]]

# Convert tokens to strings properly using `convert_tokens_to_string`.
# Note: `.convert_tokens_to_string()` expects a list of tokens.
converted_strings = [new_tokenizer.convert_tokens_to_string([t]) for t in token_strings]

print(converted_strings)
```

## Creating IterableDataset Using PyTorch

```python
import torch
from torch.utils.data import IterableDataset
from transformers import AutoTokenizer
from datasets import load_dataset
from tqdm import tqdm


# Load your custom tokenizer
model_ckpt = "Jamoni/codeparrot"  # Replace with your model checkpoint
tokenizer = AutoTokenizer.from_pretrained(model_ckpt)

# Load the dataset
dataset_name = 'transformersbook/codeparrot'
dataset = load_dataset(dataset_name, split="train", streaming=True)


class ConstantLengthDataset(IterableDataset):
    def __init__(self, tokenizer, dataset, seq_length=1024, num_of_sequences=1024, chars_per_token=3.6):
        self.tokenizer = tokenizer
        self.concat_token_id = tokenizer.eos_token_id
        self.dataset = dataset
        self.seq_length = seq_length
        self.input_characters = num_of_sequences * seq_length * chars_per_token

    def __iter__(self):
        iterator = iter(self.dataset)
        more_examples = True

        while more_examples:
            buffer, buffer_len = [], 0

            # Fill the buffer until it reaches the input character limit
            while True:
                if buffer_len >= self.input_characters:
                    print(f"Buffer full: {buffer_len} >= {self.input_characters:.0f}")
                    break  # Exit if the buffer is full

                try:
                    print(f"Fill buffer: {buffer_len} < {self.input_characters:.0f}")
                    example = next(iterator)
                    buffer.append(example['content'])
                    buffer_len += len(example['content'])
                except StopIteration:
                    more_examples = False  # Exit when there are no more items in the dataset
                    print("No more examples in dataset.")
                    break

            all_token_ids = []
            tokenized_inputs = self.tokenizer(buffer,truncation=False)

            for tokenized_input in tokenized_inputs['input_ids']:
                all_token_ids.extend(tokenized_input + [self.concat_token_id])

            # Yield input_ids in chunks of seq_length
            for i in range(0, len(all_token_ids), self.seq_length):
                input_ids = all_token_ids[i : i + self.seq_length]
                if len(input_ids) == self.seq_length:
                    yield torch.tensor(input_ids)


# Create an instance of ConstantLengthDataset
constant_length_dataset = ConstantLengthDataset(tokenizer, dataset, num_of_sequences=10)

# Create an iterator and print some sequence lengths
dataset_iterator = iter(constant_length_dataset)
lengths = [len(b) for _, b in zip(range(5), dataset_iterator)]
print(f'Sequence lengths: {lengths}')
```

```python
import gradio as gr
from transformers import AutoTokenizer

# Load your fine-tuned CodeParrot tokenizer
model_ckpt = "Jamoni/codeparrot"
tokenizer = AutoTokenizer.from_pretrained(model_ckpt)

def tokenize_code(python_code):
    """
    Tokenize the given Python code and return the predicted tokens.

    Parameters:
    python_code (str): The Python code to be tokenized.

    Returns:
    str: The predicted tokens, one per line.
    """
    # Tokenize the input code using the fine-tuned model
    tokens = tokenizer(python_code).tokens()
    return '\n'.join(tokens)

def main():
    demo = gr.Interface(
        fn=tokenize_code,
        inputs=gr.Code(label="Python Code", language="python"),
        outputs=gr.Textbox(label="Predicted Tokens"),
        title="Python Code Tokenizer",
        description="Enter Python code and see the predicted tokens."
    )
    demo.launch(inline=True)

if __name__ == "__main__":
    main()

```
